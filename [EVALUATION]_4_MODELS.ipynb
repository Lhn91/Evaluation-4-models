#CHECK DATASET chuẩn yolo và cấu trúc file

```python
import os

dataset_paths = [
    '/content/T1_C2A_ORIGINAL_10k',
    '/content/T1_C2A_ORIGINAL_WEATHER_10K',
    '/content/T2_VISDRONE',
    '/content/T3_HERIDAL',
    '/content/T3_SARD'
]

# Các định dạng ảnh phổ biến cần đếm
IMAGE_EXTS = ('.jpg', '.jpeg', '.png', '.bmp', '.webp', '.tif', '.tiff')

def print_directory_tree(base_path):
    if not os.path.exists(base_path):
        print(f"❌ Không tìm thấy: {base_path}")
        return

    # 1. Quét nhanh để tính TỔNG số lượng của toàn bộ dataset
    total_images = 0
    total_txt = 0
    for root, _, files in os.walk(base_path):
        total_images += sum(1 for f in files if f.lower().endswith(IMAGE_EXTS))
        total_txt += sum(1 for f in files if f.lower().endswith('.txt'))

    print(f"📁 {base_path}")
    print(f"📊 TỔNG CỘNG DATASET: ** {total_images} ảnh ** | ** {total_txt} file .txt **")
    print("Cấu trúc chi tiết:")

    # 2. Duyệt và in sơ đồ cây thư mục
    for root, dirs, files in os.walk(base_path):
        rel_path = os.path.relpath(root, base_path)

        # Xử lý riêng các file nằm rải rác ngay ở thư mục gốc (nếu có)
        if rel_path == '.':
            img_count = sum(1 for f in files if f.lower().endswith(IMAGE_EXTS))
            txt_count = sum(1 for f in files if f.lower().endswith('.txt'))
            if img_count > 0 or txt_count > 0:
                print(f"    ├── [Files ở gốc] : {img_count} ảnh | {txt_count} .txt")
            continue

        # Tính toán khoảng cách thụt lề cho thư mục con
        level = rel_path.count(os.sep)
        indent = '    ' * (level + 1)
        folder_name = os.path.basename(root)

        # Đếm file trong thư mục hiện tại
        img_count = sum(1 for f in files if f.lower().endswith(IMAGE_EXTS))
        txt_count = sum(1 for f in files if f.lower().endswith('.txt'))

        # In ra tên thư mục kèm số lượng
        print(f"{indent}├── 📂 {folder_name} ({img_count} ảnh | {txt_count} .txt)")

print("=" * 60)
for path in dataset_paths:
    print_directory_tree(path)
    print("-" * 60)
```

#EVALUATION

```python
import os
import yaml

def create_yaml_files():
    # 1. KHAI BÁO THƯ MỤC LƯU YAML SẼ TẠO RA
    output_yaml_dir = '/content/drive/MyDrive/YAML_Configs_5'
    os.makedirs(output_yaml_dir, exist_ok=True)

    # 2. ĐỊNH NGHĨA DANH SÁCH PATHS (CẤU TRÚC FOLDER THỰC TẾ TRÊN ẢNH)
    # Tên_Dataset: [Danh_sách_các_đường_dẫn_chứa_ảnh_cần_test_bên_trong]
    # Dùng List (danh sách) để YOLO có thể gộp nhiều folder lại khi chạy validation
    datasets_to_create = {
        # Nhóm T1: Chỉ test (in-domain)
        'T1_C2A_Original': [
            '/content/T1_C2A_ORIGINAL_10k/test/images'
        ],
        'T1_C2A_Weather': [
            '/content/T1_C2A_ORIGINAL_WEATHER_10K/test/images'
        ],

        # Nhóm T2: Seen domain (Chỉ dùng test hoặc valid, TUYỆT ĐỐI không dùng train)
        'T2_VisDrone': [
            '/content/T2_VISDRONE/test/images'
        ],

        # Nhóm T3: Unseen (Dùng TOÀN BỘ data, gộp tất cả các folder có chứa ảnh)
        'T3_Heridal': [
            '/content/T3_HERIDAL/train/images',
            '/content/T3_HERIDAL/valid/images',
            '/content/T3_HERIDAL/test/images'# Vì Heridal của bạn chỉ có folder train
        ],

        'T3_SARD': [
            '/content/T3_SARD/train/images',
            '/content/T3_SARD/valid/images',
            '/content/T3_SARD/test/images'
        ]
    }

    # Class names mặc định cho bài toán (nếu model của bạn chỉ detect 1 class là 'person')
    class_names = {0: 'person'}

    print(f"Bắt đầu tạo {len(datasets_to_create)} file YAML với cấu trúc gộp thư mục...\n")

    for dataset_name, val_folders_list in datasets_to_create.items():
        yaml_path = os.path.join(output_yaml_dir, f"{dataset_name}.yaml")

        # Tạo nội dung cho file YAML (Không dùng 'path' gốc nữa mà điền thẳng absolute path vào list)
        yaml_content = {
            'train': val_folders_list, # Fake train path (YOLO bắt buộc có)
            'val': val_folders_list,   # Khi val() chạy, nó sẽ đọc toàn bộ list đường dẫn này
            'test': val_folders_list,
            'names': class_names
        }

        # Ghi ra file với Dumper tùy chỉnh để tắt Alias
        class NoAliasDumper(yaml.SafeDumper):
            def ignore_aliases(self, data):
                return True

        with open(yaml_path, 'w', encoding='utf-8') as f:
            yaml.dump(yaml_content, f, Dumper=NoAliasDumper, sort_keys=False)

        print(f"✅ Đã tạo: {yaml_path}")
        for p in val_folders_list:
            print(f"   + Đã gộp path: {p}")
        print("-" * 30)

    print("="*50)
    print("HOÀN TẤT! Bạn có thể dùng các file YAML này để nạp vào script chạy Eval.")

if __name__ == "__main__":
    create_yaml_files()
```

```python
# Cài đặt thư viện (Chạy 1 lần trên Google Colab)
!pip install ultralytics pandas matplotlib seaborn openpyxl

import os
import yaml
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from ultralytics import YOLO

# ==========================================
# 1. KHAI BÁO ĐƯỜNG DẪN TẠI ĐÂY
# ==========================================

# Điền đường dẫn các file weights .pt của bạn trên Google Drive
models = {
    'M1_K_Original_C2A': '/content/drive/MyDrive/Projcet-51_Project-52/[result] Training process result/M1_Roboflow-Yoloc8-C2A-Original_PhuongNguyen.pt',
    'M2_R_Weather_C2A': '/content/drive/MyDrive/Projcet-51_Project-52/[result] Training process result/M2-Roboflow-Yolov8-C2A-Original-weather-10k-NguyenLe.pt',
    'M3_Mix_Base': '/content/drive/MyDrive/Projcet-51_Project-52/[result] Training process result/M3_YoloV8_Mix_C2A_no_Weather_NguyenLe.pt',
    'M4_Mix_Weather': '/content/drive/MyDrive/Projcet-51_Project-52/[result] Training process result/M4_Yolov8_Mix_C2A_Weather__BaoLong.pt'
}

# Điền đường dẫn các file .yaml cấu hình dataset của bạn
# LƯU Ý: T1, T2 file yaml CHỈ được trỏ đến thư mục test/val.
# T3 file yaml có thể trỏ đến thư mục chứa gộp tất cả ảnh (train+val+test).
datasets = {
    'T1_C2A_Original': '/content/drive/MyDrive/YAML_Configs_5/T1_C2A_Original.yaml',
    'T1_C2A_Weather': '/content/drive/MyDrive/YAML_Configs_5/T1_C2A_Weather.yaml',
    'T2_VisDrone': '/content/drive/MyDrive/YAML_Configs_5/T2_VisDrone.yaml',
    'T3_Heridal': '/content/drive/MyDrive/YAML_Configs_5/T3_Heridal.yaml',
    'T3_SARD': '/content/drive/MyDrive/YAML_Configs_5/T3_SARD.yaml'
}

output_dir = '/content/drive/MyDrive/SAR_Evaluation_Results_FINAL_3'
os.makedirs(output_dir, exist_ok=True)

# ==========================================
# 2. CHẠY EVALUATION TỰ ĐỘNG
# ==========================================

results_data = []

for dataset_name, yaml_path in datasets.items():
    # Đọc file yaml để ước tính số ảnh (tùy chọn, giả sử có key 'val')
    try:
        with open(yaml_path, 'r') as f:
            data_info = yaml.safe_load(f)
            # Tùy file yaml của bạn cấu hình key nào để chạy val
            val_path = data_info.get('val', '')
            if not os.path.isabs(val_path):
                val_path = os.path.join(data_info.get('path', ''), val_path)
            num_images = len(os.listdir(val_path)) if os.path.exists(val_path) else "N/A"
    except Exception as e:
        num_images = "N/A"

    for model_name, model_path in models.items():
        print(f"\n{'='*60}")
        print(f"🚀 Evaluating {model_name} on {dataset_name}...")
        print(f"{'='*60}")

        # Xóa bộ nhớ GPU để tránh Out of Memory trên Colab
        import torch
        torch.cuda.empty_cache()

        model = YOLO(model_path)

        # Chạy Validation.
        # Lưu ý: split='val' là mặc định của YOLO để đọc thư mục khai báo trong key 'val' của file yaml.
        metrics = model.val(
            data=yaml_path,
            conf=0.25,
            iou=0.45,
            split='val',
            save_json=True,
            project=output_dir,
            name=f'val_{model_name}_{dataset_name}',
            exist_ok=True,
            batch=8, # Tùy GPU Colab cấp (T4, V100), có thể chỉnh lên 16
            device=0
        )

        # Bóc tách chỉ số
        map50 = metrics.box.map50
        map50_95 = metrics.box.map
        precision = metrics.box.mp
        recall = metrics.box.mr
        f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

        # Cố gắng tính FPPI (nếu lấy được).
        # Cách lấy đơn giản nhất: tổng detections - true positives.
        # Tuy nhiên ultralytics không trả về trực tiếp mảng TP/FP dễ lấy. Tạm thời dùng biến giả,
        # bạn có thể tính sau hoặc trích xuất từ file kết quả JSON.
        fppi = "Chờ tính thủ công từ JSON"

        results_data.append({
            'Model': model_name,
            'Dataset': dataset_name,
            'mAP@50': round(map50, 4),
            'mAP@50-95': round(map50_95, 4),
            'Precision': round(precision, 4),
            'Recall': round(recall, 4),
            'F1-Score': round(f1, 4),
            'Num_Images': num_images
        })

# ==========================================
# 3. XUẤT EXCEL
# ==========================================
df = pd.DataFrame(results_data)
# Tạo MultiIndex
df_pivot = df.pivot(index='Model', columns='Dataset', values=['mAP@50', 'mAP@50-95', 'Precision', 'Recall', 'F1-Score'])
df_pivot.columns = df_pivot.columns.swaplevel(0, 1)
df_pivot.sort_index(axis=1, level=0, inplace=True)

excel_path = os.path.join(output_dir, 'SAR_Evaluation_Results.xlsx')
df_pivot.to_excel(excel_path)
print(f"\n✅ Đã lưu kết quả Excel tại: {excel_path}")

# ==========================================
# 4. VẼ BIỂU ĐỒ RECALL SO SÁNH TRÊN TẬP T3
# ==========================================
target_datasets = ['T3_Heridal', 'T3_SARD']
plot_data = df[df['Dataset'].isin(target_datasets)]

if not plot_data.empty:
    plt.figure(figsize=(10, 6))
    sns.barplot(data=plot_data, x='Dataset', y='Recall', hue='Model')
    plt.title('Comparison of Recall across T3 Unseen Datasets (conf=0.25)', fontsize=14)
    plt.ylabel('Recall', fontsize=12)
    plt.xlabel('Dataset', fontsize=12)
    plt.ylim(0, 1.0)
    plt.legend(title='Models', bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()

    plot_path = os.path.join(output_dir, 'recall_comparison_T3.png')
    plt.savefig(plot_path, dpi=300)
    print(f"✅ Đã lưu biểu đồ tại: {plot_path}")
    plt.show()
```
